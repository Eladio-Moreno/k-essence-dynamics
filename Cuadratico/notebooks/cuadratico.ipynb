```python
v_variable = [1e-16,1e-17,1e-18,1e-19]#,0,1e-15,1e-18, 1e-20]
m=[]
for v in v_variable:
    common_settings['m_phi_schm']=v
    m.append(Class())
    m[-1].set(common_settings)
    m[-1].compute()
```

```python
plt.figure(figsize=(7, 5))
for i in range(len(v_variable)):
    pert = m[i].raw_cl()
    tt = pert['tt']
    ell = pert['ell']
    
    plt.plot(ell,(ell*(ell +1))*tt,label=" $ m_{\\phi} = $"+str(v_variable[i]))
    
plt.title('Power spectrum', fontsize=9)
plt.xlabel('$l$',fontsize=9)
plt.ylabel('$l(l+1)C_{l} \;/\;2\pi \;$[$\; \mu k^{2}$]',fontsize=9)    
plt.legend(fontsize=9)
```
