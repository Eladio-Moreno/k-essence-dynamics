```python
# import classy module
from classy import Class
```

```python
# create instance of the class "Class"
LambdaCDM = Class()
# pass input parameters
LambdaCDM.set({'omega_b':0.022032,'omega_cdm':0.12038,'h':0.67556,'A_s':2.215e-9,'n_s':0.9619,'tau_reio':0.0925})
LambdaCDM.set({'output':'tCl,pCl,lCl,mPk','lensing':'yes','P_k_max_1/Mpc':3.0})
# run class
LambdaCDM.compute()
```

```python
# create instance of the class "Class"
cuad = Class()
# pass input parameters
cuad.set({'Omega_fld' : 0,
                   'Omega_smg' : -1,
                   'Omega_Lambda': 0.69,
                   'omega_b':0.022032,
                   'Omega_cdm':0,
                   'gravity_model':'quintessence_monomial',
                   'theta_schm':2e-17,
                   'DM_schm':0.26,
                   'h_schm':2,
                   'x0_schm':10,
                   'm_phi_schm':1e-16,
                   'h':0.6732117,
                   'A_s':2.215e-9,
                   'n_s':0.9619,
                   'tau_reio':0.0925})
cuad.set({'output':'tCl,pCl,lCl,mPk','lensing':'yes','P_k_max_1/Mpc':3.0,'k_output_values':'0.1'})
# run class
cuad.compute()
```

```python
# get all C_l output
cls = LambdaCDM.lensed_cl(2500)
# To check the format of cls
#cls.viewkeys()
```

```python
# get all C_l output
cls1 = cuad.lensed_cl(2500)
# To check the format of cls
#cls.viewkeys()
```

```python
ll = cls['ell'][2:]
clTT = cls['tt'][2:]
clEE = cls['ee'][2:]
clPP = cls['pp'][2:]
```

```python
ll1 = cls1['ell'][2:]
clTT1 = cls1['tt'][2:]
clEE1 = cls1['ee'][2:]
clPP1 = cls1['pp'][2:]
```

```python
# uncomment to get plots displayed in notebook
%matplotlib inline
import matplotlib.pyplot as plt
from math import pi
```

```python
# plot C_l^TT
plt.figure(1)
plt.xscale('log');plt.yscale('linear');plt.xlim(2,2500)
plt.xlabel(r'$\ell$')
plt.ylabel(r'$[\ell(\ell+1)/2\pi]  C_\ell^\mathrm{TT}$')
plt.plot(ll,clTT*ll*(ll+1)/2./pi,'r-',label=r"$\Lambda$CDM")
plt.plot(ll1,clTT1*ll1*(ll1+1)/2./pi,'b-', label="cuadrático")
plt.legend()
```

```python
plt.savefig('warmup_cltt.pdf')
```

```python
# get P(k) at redhsift z=0
import numpy as np
kk = np.logspace(-4,np.log10(3),1000) # k in h/Mpc
Pk = [] # P(k) in (Mpc/h)**3
h = LambdaCDM.h() # get reduced Hubble for conversions to 1/Mpc
for k in kk:
    Pk.append(LambdaCDM.pk(k*h,0.)*h**3) # function .pk(k,z)
```

```python
# get P(k) at redhsift z=0

kk = np.logspace(-4,np.log10(3),1000) # k in h/Mpc
Pk1 = [] # P(k) in (Mpc/h)**3
h1 = cuad.h() # get reduced Hubble for conversions to 1/Mpc
for k in kk:
    Pk1.append(cuad.pk(k*h1,0.)*h1**3) # function .pk(k,z)
```

```python
# plot P(k)
plt.figure(2)
plt.xscale('log');plt.yscale('log');plt.xlim(kk[0],kk[-1])
plt.xlabel(r'$k \,\,\,\, [h/\mathrm{Mpc}]$')
plt.ylabel(r'$P(k) \,\,\,\, [\mathrm{Mpc}/h]^3$')
plt.plot(kk,Pk,'b-', label=r"$\Lambda$CDM")
plt.plot(kk,Pk1,'r-',label=r"Cuadrático")
plt.legend()
```
