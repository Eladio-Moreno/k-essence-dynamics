```python
z= background['z']                    # Redshift
#H_z = background['H [1/Mpc]']          # Hubble parameter H(z)
a = 1 / (1 + z)                        # Factor de escala
# Densidades de energía
rho_gamma = background['(.)rho_g']     # Densidad de energía de fotones
rho_nu = background['(.)rho_ur']     # Densidad de energía de neutrinos
rho_rad = rho_gamma + rho_nu              # Radiación total (fotones + neutrinos relativistas)
rho_bar = background['(.)rho_b']           # Densidad de energía bariónica
rho_lambda = background['(.)rho_lambda']     # Densidad de energía de la constante cosmológica
rho_crit = background['(.)rho_crit']   # Densidad crítica
rho_dm = background['(.)rho_smg']     # Densidad de energía de materia oscura
rho_m = rho_bar + rho_dm # Densidad de energía de materia total
cs2 = background['c_s^2']
phi = background['phi_smg']
phi_prime = background["phi'"]
```
